# Bar Graph

Basic Bar Graph

```python
children=[10,20,30,40,10]
colors=['red','blue','green','yellow','pink']
plt.bar(colors,children)
```

Horizontal Bar Graph ==> Used when categorical data is large

```python
plt.barh(colors,children)
```

Multiple Bar Chart

```python
df2=pd.read_csv('batsman_season_record.csv')
```

```python
df2
```

```python
plt.bar(df2['batsman'],df2['2015'])
plt.bar(df2['batsman'],df2['2016'])
```

```python
plt.bar(np.arange(df2.shape[0])-0.2,df2['2015'],width=0.2)
plt.bar(np.arange(df2.shape[0]),df2['2016'],width=0.2)
plt.bar(np.arange(df2.shape[0])+0.2,df2['2017'],width=0.2)
```

Using Ticks changing labels of x-axis

```python
plt.bar(np.arange(df2.shape[0])-0.2,df2['2015'],width=0.2)
plt.bar(np.arange(df2.shape[0]),df2['2016'],width=0.2)
plt.bar(np.arange(df2.shape[0])+0.2,df2['2017'],width=0.2)
plt.xticks(np.arange(df2.shape[0]),df2['batsman'])
plt.show()
```

label Overlapping

```python
children=[10,20,30,40,10]
colors=['red red red red red ','blue blue blue blue ','green green green green ',' yellow yellow','pink pink pink']
plt.bar(colors,children)
```

```python
children=[10,20,30,40,10]
colors=['red red red red red ','blue blue blue blue ','green green green green ',' yellow yellow','pink pink pink']
plt.bar(colors,children)
plt.xticks(rotation='vertical')
plt.show()
```

Stacked Bar Chart

```python
plt.bar(df2['batsman'],df2['2017'],label='2017')
plt.bar(df2['batsman'],df2['2016'],bottom=df2['2017'],label='2016')
plt.bar(df2['batsman'],df2['2015'],bottom=df2['2016']+df2['2017'],label='2015')
plt.legend()
```

# Pie

Simple Data

```python
data=[23,45,100,20,49]
plt.pie(data)
plt.show()
```

Labelling

```python
data=[23,45,100,20,49]
sub=['eng','hindi','maths','science','sst']
plt.pie(data,labels=sub)
plt.show()
```

Pandas dataset

```python
hitter=pd.read_csv('gayle-175.csv')
```

```python
hitter
```

```python
plt.pie(hitter['batsman_runs'],labels=hitter['batsman'])
plt.show()
```

```python
plt.pie(hitter['batsman_runs'],labels=hitter['batsman'],autopct='%0.1f%%')
plt.show()
```

Colors

```python
plt.pie(hitter['batsman_runs'],labels=hitter['batsman'],autopct='%0.1f%%',colors=['blue','green','yellow','pink','cyan','brown'])
plt.show()
```

Explode

```python
plt.pie(hitter['batsman_runs'],labels=hitter['batsman'],autopct='%0.1f%%',explode=[0.1,0,0,0,0,0])
plt.show()
```

Shadow

```python
plt.pie(hitter['batsman_runs'],labels=hitter['batsman'],autopct='%0.1f%%',explode=[0.1,0,0,0,0,0],shadow=True)
plt.show()
```
